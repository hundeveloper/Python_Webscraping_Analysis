```python
import requests
import os
import json
from dotenv import load_dotenv

# .env íŒŒì¼ì—ì„œ í™˜ê²½ ë³€ìˆ˜ ë¡œë“œ
load_dotenv()
NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")

# ê³µí†µ í—¤ë” ì„¤ì •
headers = {
    "X-Naver-Client-Id": NAVER_CLIENT_ID,
    "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
}

# ì‡¼í•‘ ì•„ì´í…œ ì¶œë ¥ í•¨ìˆ˜
def print_shop_item(idx, item):
    title = item['title'].replace('<b>', '').replace('</b>', '')
    lprice = f"{int(item['lprice']):,}ì›"
    brand = item.get('brand', '').strip()
    mall = item.get('mallName', '').strip()
    link = item['link']

    print(f"{idx}. {title}")
    print(f"   ìµœì €ê°€: {lprice}")
    print(f"   ë¸Œëœë“œ: {brand}")
    print(f"   ì‡¼í•‘ëª°: {mall}")
    print(f"   ë§í¬: {link}")
    print()

# ì‡¼í•‘ ê²€ìƒ‰ í•¨ìˆ˜
def search_shops(query):
    url = "https://openapi.naver.com/v1/search/shop.json"
    payload = {
        "query": query,
        "display": 10,
        "sort": "sim"
    }

    res = requests.get(url, headers=headers, params=payload)
    if res.status_code == 200:
        data = res.json()
        items = data.get('items', [])
        if items:
            print("=== ì‡¼í•‘ ê²€ìƒ‰ ê²°ê³¼ ===\n")
            for idx, item in enumerate(items, 1):
                print_shop_item(idx, item)
            # JSON íŒŒì¼ë¡œ ì €ì¥
            with open("shops.json", "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=4)
            print("ğŸ”½ shops.json íŒŒì¼ë¡œ ì €ì¥ ì™„ë£Œ!")
        else:
            print("ê²€ìƒ‰ ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.")
    else:
        print(f"API í˜¸ì¶œ ì‹¤íŒ¨: {res.status_code}")
        print("ì‘ë‹µ ë©”ì‹œì§€:", res.text)

# ì˜ˆì‹œ ì‹¤í–‰
search_shops("ê°€ë””ê±´")
```

```python
#3-2 ì§ˆë¬¸

import requests
import os
from dotenv import load_dotenv

# .env íŒŒì¼ì—ì„œ í™˜ê²½ ë³€ìˆ˜ ë¡œë“œ
load_dotenv()
NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")

# ê³µí†µ í—¤ë” ì„¤ì •
headers = {
    "X-Naver-Client-Id": NAVER_CLIENT_ID,
    "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
}

# ì‡¼í•‘ ì•„ì´í…œ ì¶œë ¥ í•¨ìˆ˜
def print_shop_item(idx, item):
    title = item['title'].replace('<b>', '').replace('</b>', '')
    lprice = f"{int(item['lprice']):,}ì›"
    brand = item.get('brand', '').strip()
    mall = item.get('mallName', '').strip()
    link = item['link']

    print(f"{idx}. {title}")
    print(f"   ìµœì €ê°€: {lprice}")
    print(f"   ë¸Œëœë“œ: {brand}")
    print(f"   ì‡¼í•‘ëª°: {mall}")
    print(f"   ë§í¬: {link}")
    print()

# ê°€ê²© ì¡°ê±´ í•„í„°ë§ í›„ ì¶œë ¥
def search_shops_under_price(query, max_price):
    url = "https://openapi.naver.com/v1/search/shop.json"
    payload = {
        "query": query,
        "display": 3,
        "sort": "sim"
    }

    res = requests.get(url, headers=headers, params=payload)
    if res.status_code == 200:
        items = res.json().get('items', [])
        filtered_items = [item for item in items if int(item['lprice']) <= max_price]

        if filtered_items:
            print(f"=== {max_price:,}ì› ì´í•˜ ìƒí’ˆ ===\n")
            for idx, item in enumerate(filtered_items, 1):
                print_shop_item(idx, item)
        else:
            print(f"{max_price:,}ì› ì´í•˜ ìƒí’ˆì´ ì—†ìŠµë‹ˆë‹¤.")
    else:
        print(f"API í˜¸ì¶œ ì‹¤íŒ¨: {res.status_code}")
        print("ì‘ë‹µ ë©”ì‹œì§€:", res.text)

# ì˜ˆì‹œ ì‹¤í–‰
search_shops_under_price("ê°€ë””ê±´", 50000)
```

```python
import pandas as pd
import json

# JSON íŒŒì¼ ë¡œë”© (ì˜ˆ: shops_filtered.json ë˜ëŠ” shops.json)
with open("shops_filtered.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# DataFrame ë³€í™˜
df = pd.DataFrame(data)

mall_filter = "ë„¤ì´ë²„"
df_mall = df[df['mallName'] == mall_filter]

# lpriceë¥¼ ì •ìˆ˜í˜•ìœ¼ë¡œ ë³€í™˜ (ì •ë ¬ì„ ìœ„í•´)
df_mall.loc[:, 'lprice'] = df_mall['lprice'].astype(int)

# ì»¬ëŸ¼ ìˆœì„œ í™•ì¸
# print(df_mall.columns)  # lprice, productId, brand, mallName, link ë“±

# lpriceë¶€í„° brandê¹Œì§€ ìŠ¬ë¼ì´ì‹±í•˜ì—¬ ì„ íƒ (ìœ„ì¹˜ ê¸°ì¤€)
start_col = df_mall.columns.get_loc('lprice')
end_col = df_mall.columns.get_loc('brand')
df_selected = df_mall.iloc[:, start_col:end_col+1]

# ê°€ê²© ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬ + index ì´ˆê¸°í™”
df_sorted = df_selected.sort_values(by='lprice').reset_index(drop=True)

# ê²°ê³¼ ì¶œë ¥
print("=== íŠ¹ì • ì‡¼í•‘ëª° ìƒí’ˆ ëª©ë¡ (ê°€ê²© ì˜¤ë¦„ì°¨ìˆœ) ===\n")
print(df_sorted)

# í•„ìš” ì‹œ CSV ì €ì¥
df_sorted.to_csv("filtered_sorted_shops.csv", index=False, encoding="utf-8-sig")
```
