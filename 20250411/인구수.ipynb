```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# 한글 폰트 설정
mpl.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df = pd.read_csv('인구현황.csv', encoding='utf-8')
df.columns = df.columns.str.strip()

# 숫자형 변환
df['총인구수'] = pd.to_numeric(df['총인구수'], errors='coerce')

# '전국' 제외한 집계
pop_by_region= df.groupby('행정기관')['총인구수'].sum().sort_values(ascending=False)
pop_by_region= pop_by_region.drop('전국', errors='ignore') 

# 시각화용 DF
pop_df = pop_by_region.reset_index()
pop_df.columns = ['행정기관', '총인구수']

# 1-1. 지역별 총인구수 plot
plt.figure(figsize=(12, 6))
sns.barplot(data=pop_df, x='행정기관', y='총인구수', hue='행정기관', palette='Spectral', dodge=False, legend=False)
plt.title('지역별 총인구수 (전국 제외)')
plt.xlabel('행정기관')
plt.ylabel('총인구수')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()

# 2. 세대당 인구수 높은 지역 Top5
df['세대수'] = pd.to_numeric(df['세대수'], errors='coerce')
df = df[df['세대수'] > 0]

df['세대당 인구'] = df['총인구수'] / df['세대수']
avg_by_region = df.groupby('행정기관')['세대당 인구'].mean().sort_values(ascending=False)
top5_household = avg_by_region.head(5).reset_index()

# 표 출력
print("세대당 인구수 높은 지역 Top5")
print(top5_household)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# 한글 폰트 설정
mpl.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df = pd.read_csv('인구현황.csv', encoding='utf-8-sig')
df.columns = df.columns.str.strip()

# '전국' 제거
df = df[df['행정기관'] != '전국']

# 쉼표 제거 후 숫자형으로 변환
for col in ['남자 인구수', '여자 인구수', '총인구수']:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')

# 남녀 비율 계산
df['남여 비율'] = round(df['남자 인구수'] / df['여자 인구수'], 2)

# 정렬
sorted_df = df[['행정기관', '남여 비율']].sort_values(by='남여 비율', ascending=False).reset_index(drop=True)

# 시각화 - 세로 막대그래프
plt.figure(figsize=(12, 6))
sns.barplot(data=sorted_df, x='행정기관', y='남여 비율', palette='coolwarm', hue='행정기관', legend=False)

# 기준선 추가
plt.axhline(y=1, color='red', linestyle='--', label='남녀 동일 비율 (1.0)')

# 제목 및 라벨
plt.title('행정기관별 남녀 비율 분석 (세로 막대그래프)')
plt.ylabel('남녀 비율 (남자 / 여자)')
plt.xlabel('행정기관')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 남초/여초 여부를 새로운 컬럼으로 추가
sorted_df['남녀 구분'] = sorted_df['남여 비율'].apply(lambda x: '남초' if x > 1 else ('여초' if x < 1 else '비슷'))

# 시각화
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(data=sorted_df, x='행정기관', y='남여 비율', hue='남녀 구분', palette={'남초': 'skyblue', '여초': 'lightcoral', '비슷': 'gray'})

# 기준선 추가
plt.axhline(1.0, color='red', linestyle='--', label='남녀 비율 1.00 기준선')

# 그래프 세부 설정
plt.title('지역별 남녀 비율 (남초 vs 여초)')
plt.xlabel('행정기관')
plt.ylabel('남여 비율 (남자 / 여자)')
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()
```
