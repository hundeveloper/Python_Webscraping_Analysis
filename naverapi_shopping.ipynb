```python
import requests
import os
import re
from dotenv import load_dotenv

# .env 파일에서 환경 변수 로드
load_dotenv()
NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")

# 공통 헤더
headers = {
    "X-Naver-Client-Id": NAVER_CLIENT_ID,
    "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
}

# 쇼핑 검색 결과 출력 함수
def print_shop_item(idx, item):
    title = re.sub(r"<.*?>", "", item.get("title", ""))
    lprice = int(item.get("lprice", 0))
    brand = item.get("brand", "").strip()
    mall = item.get("mallName", "").strip()
    link = item.get("link", "").strip()

    print(f"{idx}. {title}")
    print(f"   최저가: {lprice:,}원")
    print(f"   브랜드: {brand}")
    print(f"   쇼핑몰: {mall}")
    print(f"   링크: {link}\n")

# 특정 쇼핑몰 상품만 검색
def search_shops_by_mall(query, mall_name):
    url = "https://openapi.naver.com/v1/search/shop.json"
    all_filtered_items = []
    display = 30
    start = 1
    total_to_fetch = 30  # 최대 200개까지 조회 (2페이지)

    while start < total_to_fetch:
        payload = {
            "query": query,
            "display": display,
            "start": start,
            "sort": "sim"
        }

        res = requests.get(url, headers=headers, params=payload)
        if res.status_code != 200:
            print(f"API 호출 실패: {res.status_code}")
            break

        items = res.json().get("items", [])
        if not items:
            break

        # 쇼핑몰 필터링
        filtered = [item for item in items if item.get("mallName", "").strip() == mall_name]
        all_filtered_items.extend(filtered)

        start += display

    # 결과 출력
    if all_filtered_items:
        print(f"=== {mall_name} 상품 목록 ({len(all_filtered_items)}개) ===\n")
        for idx, item in enumerate(all_filtered_items, 1):
            print_shop_item(idx, item)
    else:
        print(f"{mall_name} 상품이 없습니다.")

# 사용 예시
search_shops_by_mall("가디건", "네이버")
```
